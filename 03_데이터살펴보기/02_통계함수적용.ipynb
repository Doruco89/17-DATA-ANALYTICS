```python
# 평균값

# 모든 열의 평균 값 : DataFrame객체.mean()
# 특정 열의 평균 값 : DataFreme객체["열 이름"].mean() / Series

import pandas as pd
```

```python
df = pd.read_csv("./data/auto-mpg.csv", header=None)

df.columns = [
    "mpg", "cylinders", "displacement", "horsepower", "weight", "acceletation", "model year", "origin", "name"
]

df.head()
```

```python
df.info()
```

```python
df.mean(numeric_only=True)
```

```python
df["mpg"].mean().round(2)
```

```python
round(df["mpg"].mean(), 1)
```

```python
# 중앙값
# 모든 열의 중앙 값 : DataFrame객체.median()
# 특정 열의 중앙 값 : DataFrame객체["열 이름"].median()

print(df.median(numeric_only=True))
print(df.mpg.median())
```
