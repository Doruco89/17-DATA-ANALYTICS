```python
# 배열 속성 / 1차원
arr1 = np.array([1,2,3,4])
arr1.shape
```

```python
# 2차원 배열 속성

arr2 = np.array([
    [1,2,3],
    [4,5,6]
])

arr2.shape
```

```python
# 3차원 속성

arr3 = np.array([
    [[1, 2],
     [3, 4]],
     [[5, 6],
      [7, 8]]
])

arr3.shape
```
