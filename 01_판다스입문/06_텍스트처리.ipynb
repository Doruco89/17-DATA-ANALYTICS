```python
# 문자열 메서드
# 문자열 매서드 적용(lower):Series 객체.str.lower() or Index 객체.str.lower()

import pandas as pd
import numpy as np

ser = pd.Series([
    "Apple_사과",
    "Banana_바나나",
    "Cherry_체리",
    np.nan],
    index=["First ", " Second", " Third", "Forth"])

ser2 = pd.Series([
    "Apple_사과",
    "Banana_바나나",
    "Cherry_체리",
    np.nan],
    index=["First ", " Second", " Third", "Forth"], dtype="string")
```

```python
print(ser.str.lower())
print(ser.str.upper())
print(ser.str.len())
```

```python
ser3 = pd.Series(["a1", "b2", "c3", "d4a5", "e6e7", "f8"], dtype="string")
print(ser3)
```

```python
print(ser3.str.extractall(r"(\d)"))

print(ser3.str.extractall(r"(\d)").groupby(level=0).agg("".join))
```

```python
ser3.str.extractall(r"([a-z])").groupby(level=0).agg("".join)
```

```python
#숫자를 별도의 컬럼으로 분리
ser3.str.extract(r"(?P<Letter>[a-z]+)(?P<Digit>\d+)")
```
