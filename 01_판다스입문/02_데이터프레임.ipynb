```python
import pandas as pd

matrix = [
    [15, "남", "덕영중"],
    [17, "여", "수리중"]
    ]

df = pd.DataFrame(matrix, index=["준서", "예은"], columns=["나이", "성별", "학교"])
print(df)
```

```python
df = df.rename(columns ={
    "나이": "연령",
    "성별": "남녀",
    "학교": "소속"
})
print(df)
```

```python
df = df.rename(index = {
    "준서": "학생1",
    "예은": "학생2"
})
print(df)
```

```python
# 행과 열의 위치 바꾸기(많이 쓰임 / 행과 열의 축을 바꿈)
# 행, 열 바꾸기 : DataFrame 객체.transpose() or DataFrame 객체.T

import pandas as pd

exam_data = {
    "이름": ["서준", "우현", "인아"],
    "수학": [90, 80, 70],
    "영어": [98, 89, 95],
    "음악": [85, 95, 100],
    "체육": [100, 90, 90]
    }

df = pd.DataFrame(exam_data)
df
```

```python
df = df.transpose()
df
```

```python
df = df.T
df
```
