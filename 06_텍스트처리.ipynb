```python
# 문자열 메서드
# 문자열 매서드 적용(lower):Series 객체.str.lower() or Index 객체.str.lower()

import pandas as pd
import numpy as np

ser = pd.Series([
    "Apple_사과",
    "Banana_바나나",
    "Cherry_체리",
    np.nan],
    index=["First ", " Second", " Third", "Forth"])

ser2 = pd.Series([
    "Apple_사과",
    "Banana_바나나",
    "Cherry_체리",
    np.nan],
    index=["First ", " Second", " Third", "Forth"], dtype="string")
```

```python
print(ser.str.lower())
print(ser.str.upper())
print(ser.str.len())
```
